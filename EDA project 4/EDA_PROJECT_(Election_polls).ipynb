## 1.What is the average sample size across all polls?


```python
import pandas as pd
data =pd.read_csv("generic_ballot_polls.csv")
average_sample_size = data['sample_size'].mean()
print(f"Average sample size is {average_sample_size}")
```

```python
import matplotlib.pyplot as plt
plt.bar(['Average Sample Size'], [data['sample_size'].mean()], color='skyblue')
plt.ylabel('Sample Size')
plt.title('Average Sample Size Across All Polls')
plt.show()
```

## 2.which pollster has the highest numbers of poll?

```python
# Count the number of polls for each pollster
pollster_poll_counts = data['pollster'].dropna().value_counts()
# Get the pollster with the highest number of polls
most_polled_pollster = pollster_poll_counts.idxmax()
highest_pollster_count = pollster_poll_counts.max()

print(f"The pollster with the highest number of polls = '{most_polled_pollster}' with {highest_pollster_count} polls.")

```

```python
sns.barplot(x=data['pollster'].value_counts().index, y=data['pollster'].value_counts().values, palette='Blues_d')
plt.xticks(rotation=90)
plt.show()
```

## 3.What are the top 5 pollsters with the highest average pollscore?"

```python
top_pollsters = data.groupby('pollster')['pollscore'].mean()
top_5_pollsters = top_pollsters.sort_values(ascending=False).head(5)
print(top_5_pollsters)
```

```python
sns.barplot(x=data.groupby('pollster')['pollscore'].mean().sort_values(ascending=False).head(5).index,
            y=data.groupby('pollster')['pollscore'].mean().sort_values(ascending=False).head(5).values, palette='Blues_d')
plt.title('Top 5 Pollsters with the Highest Average Poll Score')
plt.xticks(rotation=45)
plt.show()
```

## 4.Which election cycle has the highest average pollscore?

```python
# Calculate the average poll score for each election cycle
avg_pollscore_by_cycle = data.groupby('cycle')['pollscore'].mean()

# Find the election cycle with the highest average poll score
highest_avg_pollscore_cycle = avg_pollscore_by_cycle.idxmax()

# Display the result
print(f"The election cycle with the highest average poll score is: {highest_avg_pollscore_cycle}")
```

## 5.What is the first eight average polling duration by pollster?

```python
# Ensure 'start_date' and 'end_date' are in datetime format
data['start_date'] = pd.to_datetime(data['start_date'])
data['end_date'] = pd.to_datetime(data['end_date'])

# Calculate the polling duration in days
data['duration'] = (data['end_date'] - data['start_date']).dt.days

# Calculate the average duration by pollster and sort it
avg_duration = data.groupby('pollster')['duration'].mean().reset_index()

# Sort by the duration and select the first 8 pollsters
top_8_avg_duration = avg_duration.sort_values('duration').head(8)

# Display the first 8 average polling durations by pollster
print(top_8_avg_duration)

```

```python
# Calculate duration and average by pollster
data['duration'] = (pd.to_datetime(data['end_date']) - pd.to_datetime(data['start_date'])).dt.days
avg_duration = data.groupby('pollster')['duration'].mean().reset_index()

# Sort by duration and select the first 5
top_8_avg_duration = avg_duration.sort_values('duration').head(8)

# Plot
sns.barplot(x='pollster', y='duration', data=top_8_avg_duration, palette='Set2')
plt.title('Top 8 Pollsters with the Shortest Average Polling Duration')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 6.How many polls have a transparency score available?


```python
transparency_score_count = data['transparency_score'].notnull().sum()
print(f"Number of polls with a transparency score available: {transparency_score_count}")
```

```python
# Data for the bar plot: count of available and missing transparency scores
transparency_data = {
    "Available": data['transparency_score'].notnull().sum(),
    "Missing": data['transparency_score'].isnull().sum()
}
# Create a bar plot
plt.bar(transparency_data.keys(), transparency_data.values(), color=['green', 'red'])
plt.title("Transparency Score Availability")
plt.ylabel("Number of Polls")
plt.show()
```

## 7.Which sponsors are associated with the highest pollscore?


```python
# Group by sponsors and calculate the mean pollscore for each
# Filter out null sponsors and pollscores for accurate results
sponsors_with_pollscore = data.dropna(subset=['sponsors', 'pollscore'])
highest_pollscore_sponsors = (
    sponsors_with_pollscore.groupby('sponsors')['pollscore']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)
highest_pollscore_sponsors
```

```python
import matplotlib.pyplot as plt
# Filter out rows with missing values for sponsors or pollscore
sponsors_with_pollscore = data.dropna(subset=['sponsors', 'pollscore'])
# Group by sponsors, calculate the mean pollscore, and select the top 5 sponsors
top_sponsors = (
    sponsors_with_pollscore.groupby('sponsors')['pollscore']
    .mean()
    .sort_values(ascending=False)
    .head(5)
)
# Plot a pie chart for the top sponsors
plt.figure(figsize=(8, 8))
plt.pie(top_sponsors,labels=top_sponsors.index,autopct='%1.1f%%',startangle=140,colors=plt.cm.Paired.colors)
plt.title("Top Sponsors by Average Pollscore")
plt.show()
```

## 8.How many polls are tracking polls?

## Summary

##### key columns:
##### 1.poll_id : Unique identifier for polls.
##### 2.pollster : Name of the polling organization.
##### 3.Sponsors : Organization sponsoring the polls(eg.Economist)

##### 8.dem, rep, and ind: Percentages for Democratic, Republican, and Independent responses.
